#IMPORTS

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
```

#IMPORTING DATASET

```python
url = 'https://raw.githubusercontent.com/4GeeksAcademy/regularized-linear-regression-project-tutorial/main/demographic_health_data.csv'
df = pd.read_csv(url)


warnings.filterwarnings('ignore')
```

```python
target_column = 'Obesity_prevalence'

# Define logical groupings to reduce repetition
demographic_columns = ['TOT_POP', 'White-alone pop', 'Black-alone pop', 'Asian-alone pop']
economic_columns = ['Median_Household_Income_2018', 'Unemployment_rate_2018']
health_resource_columns = [
    'Active Physicians per 100000 Population 2018 (AAMC)',
    'Total nurse practitioners (2019)',
    'Total Hospitals (2019)',
    'ICU Beds_x'
]
# Combine all selected columns into one list
features_columns = demographic_columns + economic_columns + health_resource_columns
```

```python
features = df[features_columns]
target = df[target_column]
```

```python
df = df.dropna(subset=[target_column])
```

```python
features = features.dropna()
target = target.loc[features.index]

print(features.isnull().sum())
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Linear Regression Model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predictions
y_pred = lr.predict(X_test)

# Evaluation
print("Linear Regression R^2:", r2_score(y_test, y_pred))
print("Linear Regression RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

#EVALUATE LASSO REGRESSION

```python
# Lasso Regression Model
lasso = Lasso(alpha=1.0)
lasso.fit(X_train, y_train)

# Predictions
y_pred_lasso = lasso.predict(X_test)

# Evaluation
print("Lasso Regression R^2:", r2_score(y_test, y_pred_lasso))
print("Lasso Regression RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
```

```python
#Varying the alpha parameter
alphas = np.linspace(0, 20, 100)
r2_scores = []

for alpha in alphas:
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_train, y_train)
    y_pred_lasso = lasso.predict(X_test)
    r2_scores.append(r2_score(y_test, y_pred_lasso))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(alphas, r2_scores, marker='o')
plt.xlabel('Alpha')
plt.ylabel('R^2 Score')
plt.title('Lasso Regression R^2 Score vs Alpha')
plt.grid(True)
plt.show()
```
